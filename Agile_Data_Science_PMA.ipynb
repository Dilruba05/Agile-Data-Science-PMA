```python
# ==============================================================
# SECTION 10: DATA QUALITY ASSESSMENT
# STEP 1: VERIFY DATA TYPES
# ==============================================================
# Purpose:
# Display the data type of each variable to ensure that numerical
# and categorical variables are correctly recognised before
# preprocessing and modelling.
# ==============================================================

print("Data Types of Variables")

display(df.dtypes.to_frame(name="Data Type"))
```

```python
# ==============================================================
# SECTION 13: DATA PREPROCESSING
# STEP 1: REMOVE DUPLICATE RECORDS
# ==============================================================

print("Dataset Shape Before Removing Duplicates:")
print(df.shape)

# Remove duplicate records
df = df.drop_duplicates()

print("\nDataset Shape After Removing Duplicates:")
print(df.shape)
```

```python
# ==============================================================
# SECTION 14: FEATURE ENGINEERING
# STEP 1: LABEL ENCODING
# ==============================================================
# Purpose:
# Convert the categorical variable (CourseCategory) into numerical
# values so that it can be processed by machine learning algorithms.
# ==============================================================

from sklearn.preprocessing import LabelEncoder

# Create LabelEncoder object
label_encoder = LabelEncoder()

# Encode CourseCategory
df["CourseCategory"] = label_encoder.fit_transform(df["CourseCategory"])

# Display encoded values
print("Encoded Course Categories")

display(df[["CourseCategory"]].head(10))

print("\nCategory Mapping")

mapping = dict(zip(label_encoder.classes_,
                   label_encoder.transform(label_encoder.classes_)))

print(mapping)
```

```python
# ==============================================================
# SECTION 15: FEATURE ENGINEERING
# STEP 2: FEATURE SCALING (STANDARDIZATION)
# ==============================================================
# Purpose:
# Standardize numerical variables so that they have a mean of
# approximately 0 and a standard deviation of 1. Feature scaling
# improves model performance by ensuring that variables with
# larger ranges do not dominate the learning process.
# ==============================================================

from sklearn.preprocessing import StandardScaler

# Select numerical features
numerical_features = [
    "TimeSpentOnCourse",
    "NumberOfVideosWatched",
    "NumberOfQuizzesTaken",
    "QuizScores",
    "CompletionRate"
]

# Create StandardScaler object
scaler = StandardScaler()

# Create scaled features
scaled_features = scaler.fit_transform(df[numerical_features])

# Convert to DataFrame
scaled_df = pd.DataFrame(
    scaled_features,
    columns=[feature + "_Scaled" for feature in numerical_features]
)

# Combine original dataset with scaled features
df = pd.concat([df.reset_index(drop=True), scaled_df], axis=1)

# Display scaled features
print("Scaled Numerical Features")

display(df[[
    "TimeSpentOnCourse_Scaled",
    "NumberOfVideosWatched_Scaled",
    "NumberOfQuizzesTaken_Scaled",
    "QuizScores_Scaled",
    "CompletionRate_Scaled"
]].head(10))
```

```python
# ==============================================================
# SECTION 16: DATA PREPARATION FOR MACHINE LEARNING
# STEP 1: TRAIN-TEST SPLIT
# ==============================================================
# Purpose:
# Separate the dataset into training and testing sets.
# The training set is used to build the machine learning model,
# while the testing set is used to evaluate its predictive
# performance on unseen data.
# ==============================================================

from sklearn.model_selection import train_test_split

# Define predictor variables (X)
X = df[[
    "CourseCategory",
    "TimeSpentOnCourse_Scaled",
    "NumberOfVideosWatched_Scaled",
    "NumberOfQuizzesTaken_Scaled",
    "QuizScores_Scaled",
    "CompletionRate_Scaled"
]]

# Define target variable (y)
y = df["CourseCompletion"]

# Split dataset (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

# Display dataset sizes
print("Training Set Size:", X_train.shape)
print("Testing Set Size :", X_test.shape)
```

```python
# ==============================================================
# SECTION 17: BASELINE MACHINE LEARNING MODEL
# STEP 2: LOGISTIC REGRESSION
# ==============================================================
# Purpose:
# Train a Logistic Regression model as the baseline classifier
# for predicting course completion.
# ==============================================================

from sklearn.linear_model import LogisticRegression

# Create Logistic Regression model
baseline_model = LogisticRegression(random_state=42)

# Train model
baseline_model.fit(X_train, y_train)

print("Baseline Logistic Regression model trained successfully.")
```

```python
# ==============================================================
# SECTION 18: MODEL PREDICTION
# ==============================================================
# Purpose:
# Generate predictions for the testing dataset using the trained
# baseline model.
# ==============================================================

# Predict test data
y_pred = baseline_model.predict(X_test)

print("Predictions generated successfully.")
```

```python
# ==============================================================
# SECTION 19: MODEL EVALUATION
# ==============================================================
# Purpose:
# Evaluate the predictive performance of the baseline model
# using standard classification metrics.
# ==============================================================

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Baseline Model Performance")
print("-" * 35)
print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-Score : {f1:.4f}")

print("\nClassification Report")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred))
```

```python
# ==============================================================
# SECTION 20: CONFUSION MATRIX VISUALIZATION
# ==============================================================
# Purpose:
# Visualize the confusion matrix of the baseline Logistic
# Regression model.
# ==============================================================

import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_pred,
    cmap="Blues",
    values_format="d"
)

plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

```python
# ==============================================================
# SECTION 21: IMPROVED MACHINE LEARNING MODEL
# STEP 1: RANDOM FOREST CLASSIFIER
# ==============================================================
# Purpose:
# Improve the baseline Logistic Regression model by using a
# Random Forest Classifier, which can capture more complex
# relationships between learner behaviours and course completion.
# ==============================================================

from sklearn.ensemble import RandomForestClassifier

# Create Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

# Train model
rf_model.fit(X_train, y_train)

print("Random Forest model trained successfully.")
```

```python
# ==============================================================
# SECTION 22: MODEL PREDICTION
# ==============================================================

rf_pred = rf_model.predict(X_test)

print("Predictions generated successfully.")
```

```python
# ==============================================================
# SECTION 23: MODEL EVALUATION
# ==============================================================

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)

rf_accuracy = accuracy_score(y_test, rf_pred)
rf_precision = precision_score(y_test, rf_pred)
rf_recall = recall_score(y_test, rf_pred)
rf_f1 = f1_score(y_test, rf_pred)

print("="*50)
print("RANDOM FOREST PERFORMANCE")
print("="*50)

print(f"Accuracy : {rf_accuracy:.4f}")
print(f"Precision: {rf_precision:.4f}")
print(f"Recall   : {rf_recall:.4f}")
print(f"F1-Score : {rf_f1:.4f}")

print("\nClassification Report")
print(classification_report(y_test, rf_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, rf_pred))
```
